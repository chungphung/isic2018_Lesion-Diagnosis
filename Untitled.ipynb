```python
import numpy as np
for col in df.columns[1:]:
    tmp = np.ones([450, 600, 3])*255
    img_list = df['image'].loc[df[col]==1]
    weight = 1/len(img_list)
    for i in img_list:
        img = cv2.imread(join(task3_img, i+'.jpg'))
        tmp += weight*img 
    print(col)
    plt.imshow(cv2.cvtColor(tmp.astype(np.uint8), cv2.COLOR_BGR2RGB))
    plt.show()
```
